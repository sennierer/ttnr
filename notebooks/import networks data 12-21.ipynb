```python
import pandas as pd
import requests
from apis_core.helper_functions.RDFParser import RDFParser
import reversion
from django.contrib.auth.models import User
```

```python
def search_for_tt(tt):
    url = "https://lobid.org/gnd/search"
    q = {"filter": "type:CorporateBody", "q": tt}
    res = requests.get(url, params=q, headers={"accept": "application/json"})
    return res.json()
```

```python
me = User.objects.get(username="sennierer")
```

```python
memv, created = InstitutionInstitutionRelation.objects.get_or_create(name="is member of")
tt_net_v, created = InstitutionType.objects.get_or_create(name="Think Tank Network")
with reversion.create_revision():
    for idx, row in df_tts.iterrows():
        inst = Institution.objects.get(source__orig_id=row["org_id"])
        if isinstance(row["netzwerke"], str):
            lst_net = [x.strip() for x in row["netzwerke"].split(";")]
            for net in lst_net:
                if net in ["Think Tank", "Public Health", "Neue Rechte", 
                           "Behavioural economics in public governance", "Paleoliberal think tanks", 
                           "former Atlas Network", "Think Tanks in Austria (case study)", "former RELIAL",
                          "Climate Change Skeptics", "Austerity politics"]:
                    continue
                Inst2 = Institution.objects.filter(name=net)
                if Inst2.count() == 1:
                    if Inst2[0].uri_set.filter(uri__contains="d-nb.info").count() == 0:
                        tt3 = search_for_tt(f'"{Inst2[0].name}"')
                        if len(tt3['member']) == 1:
                            Uri.objects.create(uri=tt3['member'][0]['id'], entity_id=Inst2[0].id)
                    Inst2 = Inst2[0]
                elif Inst2.count() == 0:
                    tt3 = search_for_tt(f'"{net}"')
                    if len(tt3['member']) == 1:
                        Inst2 = RDFParser(tt3['member'][0]['id'], "Institution")
                        Inst2 = Inst2.get_or_create()
                    else:
                        if len(tt3['member']) > 1:
                            print(f"got several GND hits: {net}")
                        else:
                            print(f"didnt find GND hits: {net}")
                        Inst2 = Institution.objects.create(name=net)
                    Inst2.kind = tt_net_v
                    Inst2.save()
                else:
                    Inst2 = False
                    print(f"found too many hits: {net}")
                if Inst2:
                    InstitutionInstitution.objects.create(related_institutionA=inst, related_institutionB=Inst2, relation_type=memv)
    reversion.set_user(me)
    reversion.set_comment("added think tank networks from original wiki")
```
